### 효율적 시장 가설(EMH, Efficient Market Hypothesis)
> 시장에서 모든 정보가 이미 반영되어 있기 때문에 주가에는 예측이 불가능하다는 가설

#### 검정 방법
주식 시장의 주가 데이터가 랜덤 워크(Random Walk)를 따르는지 여부를 확인하는 것입니다.

랜덤 워크는 현재의 주가가 이전의 주가와 상관 없이 무작위로 결정된다는 가정을 기반으로 합니다.

주가 데이터가 랜덤 워크를 따른다면, 효율적 시장 가설이 지지될 수 있습니다.

여기서는 Augmented Dickey-Fuller (ADF) 검정을 사용하여 주가 데이터가 랜덤 워크를 따르는지를 확인하는 방법을 설명하겠습니다.

주가 데이터가 효율적 시장을 따르는지를 검정

- 랜덤 워크를 따르는 주가 데이터를 생성하고, Augmented Dickey-Fuller 검정을 통해 주가 데이터가 랜덤 워크를 따르는지를 확인합니다. 
- p-value가 유의수준인 0.05보다 작으면 귀무가설을 기각하고, 효율적 시장 가설을 기각할 수 있습니다.

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

import yfinance

# ADF 검정 함수 정의
def adf_test(series):
    result = adfuller(series)
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    print("Critical Values:", result[4])

    if result[1] <= 0.05:
        print("결과: 주가 데이터는 랜덤 워크를 따르지 않음 (효율적 시장 가설 기각)")
    else:
        print("결과: 주가 데이터는 랜덤 워크를 따를 가능성이 있음 (효율적 시장 가설 지지)")

```

```python
returns = yfinance.download('^KS11')['Adj Close']

# ADF 검정 수행
adf_test(returns)

# 주가 데이터 시각화
plt.plot(returns)
plt.title("Random Walk")
plt.xlabel("Time")
plt.ylabel("Returns")
plt.show()
```
