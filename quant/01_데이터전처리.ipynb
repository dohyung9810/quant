```python
# "티커" 컬럼 제거
data = data[[
    '시가총액', 'year', '회사명', '거래소코드', '회계년도',
       'PER(최고)(IFRS)', 'PER(최저)(IFRS)', 'PBR(최고)(IFRS)', 'PBR(최저)(IFRS)',
       'PCR(최고)(IFRS)', 'PCR(최저)(IFRS)', 'PSR(최고)(IFRS)', 'PSR(최저)(IFRS)',
       '1주당순이익(IFRS)(원)', '유동자산(*)(요약)(IFRS연결)(백만원)', '부채(*)(요약)(IFRS연결)(백만원)',
       '당기순이익(손실)(요약)(IFRS연결)(백만원)', '매출액(수익)(요약)(IFRS연결)(백만원)',
       '매출총이익(손실)(요약)(IFRS연결)(백만원)', '계속영업이익(손실)(요약)(IFRS연결)(백만원)'
]]
data.columns
```

(전처리) 시가총액 단위 원 -> 백만원

```python
# 시가총액 단위 : 원 -> 백만원
data['시가총액'] = data['시가총액'] / 1000000
```

컬럼명 변경

```python
# 컬럼명 변경
# '시가총액' -> '시가총액(백만원)' 변경
# (*), (요약), (손실), (IFRS) 제거

data.columns = [
    '시가총액(백만원)', 'year', '회사명', '거래소코드', '회계년도', 
     'PER(최고)', 'PER(최저)', 
     'PBR(최고)', 'PBR(최저)',
     'PCR(최고)', 'PCR(최저)', 
     'PSR(최고)', 'PSR(최저)', 
     '1주당순이익(원)','유동자산(백만원)', '부채(백만원)',
       '당기순이익(백만원)', '매출액(백만원)',
       '매출총이익(백만원)', '계속영업이익(백만원)'
]
```

```python
# 공통 columns
columns = ['year', '회사명', '거래소코드', '회계년도', '시가총액(백만원)']
# 시가총액(백만원) : 포트폴리오 자산 배분 시 필요

# 가치주 NCAV 전략 columns
columns_value_ncav = ['유동자산(백만원)', '부채(백만원)', '당기순이익(백만원)'] # 가치주(NCAV)
# columns_value_ncav = ['시가총액(백만원)', '유동자산(백만원)', '부채(백만원)', '당기순이익(백만원)'] # 가치주(NCAV)

# 가치주 저PER, ... 전략 columns
columns_value_low = ['PER(최고)', 'PER(최저)', 'PBR(최고)', 'PBR(최저)', 'PCR(최고)', 'PCR(최저)', 'PSR(최고)', 'PSR(최저)'] # 가치주(저PER, 저PBR, 저PCR)

# 성장주 PEG 전략 columns
columns_growth = ['PER(최고)', 'PER(최저)', '1주당순이익(원)'] # 성장주(PEG)

# 그 외(미사용)
columns_etc = ['매출액(백만원)', '매출총이익(백만원)', '계속영업이익(백만원)']
```

```python
# 전략별 컬럼 추출

# 가치주 - NCAV 전략 데이터
data_value_ncav = data[columns + columns_value_ncav]
print('NCAV 데이터 :', data_value_ncav.shape)

# 가치주 - LOW 전략 데이터
data_value_low = data[columns + columns_value_low]
print('LOW 데이터 :', data_value_low.shape)

# 성장주 - PEG 전략 데이터
data_growth = data[columns + columns_growth]
print('PEG 데이터 :', data_growth.shape)
```
