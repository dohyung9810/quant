```python
columns = ['year', '회사명', '거래소코드', '회계년도']
columns_value_ncav = ['시가총액(백만원)', '유동자산(백만원)', '부채(백만원)', '당기순이익(백만원)'] # 가치주(NCAV)
columns_value_low = ['PER(최고)', 'PER(최저)', 'PBR(최고)', 'PBR(최저)', 'PCR(최고)', 'PCR(최저)', 'PSR(최고)', 'PSR(최저)'] # 가치주(저PER, 저PBR, 저PCR)
columns_growth = ['PER(최고)', 'PER(최저)', '1주당순이익(원)'] # 성장주(PEG)

columns_etc = ['매출액(백만원)', '매출총이익(백만원)', '계속영업이익(백만원)']
```

```python
# 가치주 - NCAV 전략 데이터
data_value_ncav = data[columns + columns_value_ncav]

# 가치주 - LOW 전략 데이터
data_value_low = data[columns + columns_value_low]

# 성장주 - PEG 전략 데이터
data_growth = data[columns + columns_growth]
```

### (가치주) - NCAV

- 데이터프레임명 : `data_value_ncav`

```python
data_value_ncav
```

```python
data_value_ncav['NCAV'] = data_value_ncav['유동자산(백만원)'] - data_value_ncav['부채(백만원)']

data_value_ncav['NCAV/시총'] = data_value_ncav['NCAV'] / data_value_ncav['시가총액(백만원)']

# 조건1: ncav > 1.5*시총
data_value_ncav['조건1'] = data_value_ncav['NCAV/시총'] > 1

# 조건2: 흑자기업(당기순이익>0)
data_value_ncav['조건2'] = data_value_ncav['당기순이익(백만원)'] > 0

data_value_ncav
```

```python
flag1 = data_value_ncav['조건1']
flag2 = flag1 & data_value_ncav['조건2']

data_value_ncav[flag1]['year'].value_counts()
```

```python
flag2 = flag1 & data_value_ncav['조건2'] # ncav, 흑자기업

data_value_ncav[flag2]['year'].value_counts()
```
