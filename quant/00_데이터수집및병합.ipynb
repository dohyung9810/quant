## 데이터 병합

```python
import pandas as pd
```

```python
total_cap = pd.read_csv('datasets/kospi_cap_total.csv')
total_cap
```

```python
total_cap.info()
```

```python
# total_data = pd.read_csv('datasets/data.csv', encoding='cp949', dtypes = {'거래소코드':'object'})
total_data = pd.read_csv('datasets/data.csv', encoding='cp949')
total_data
```

```python
total_data['거래소코드'] = total_data['거래소코드'].apply(lambda x: str(x).zfill(6))
total_data.info()
```

```python
# total_data['year'] = total_data['회계년도'].split
total_data['year'] = total_data['회계년도'].apply(lambda x: x.split('/')[0]).astype('int') + 1
total_data
```

```python
merged_df = pd.merge(total_cap[['시가총액', '티커', 'year']], total_data, 
                     left_on=['티커', 'year'], 
                     right_on=['거래소코드', 'year']
                     )


merged_df.to_csv('datasets/kospi_total.csv')
```
