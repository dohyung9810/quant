```python
from pykrx import stock
```

```python
import yfinance as yf

kospi = yf.download('^KS11')
kospi
```

```python
kospi = kospi.loc['2000':]
kospi.head()
```

```python
kospi['month'] = kospi.index.month
kospi['year'] = kospi.index.year
```

```python
# kospi_last = kospi.resample('Y').last()
# kospi_last

# 연말 시총

dates = []
for year in range(2000, 2024):
    date = kospi[kospi['year']==year].tail(1).index[0]
    dates.append(date)

dates
```

```python
from tqdm import tqdm
dfs = []

for date in tqdm(dates):
    df = stock.get_market_cap_by_ticker(date=date, market='KOSPI')
    df = df.reset_index()
    df['year'] = date.year
    dfs.append(df)
```

```python
import pandas as pd
total = pd.concat(dfs)
total
```
