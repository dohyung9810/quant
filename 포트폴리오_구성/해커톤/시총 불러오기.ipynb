```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from pykrx import stock
from pykrx import bond
from tqdm import tqdm
```

```python
# 코스피 전 종목 심볼 가져오기
ks_symbol = pd.read_csv("C:/Users/USER/Desktop/해커톤/datasets/kospi전종목코드.csv", encoding='cp949')
```

```python
# ks_symbol에서는 회사명과 종목 코드만 필요하기 해당 두 열만 추출
ks_symbol = ks_symbol.iloc[1:][['회사', 'amount']]
```

```python
ks_symbol['amount']
```

```python
def bring_fundamental_info(ks_list):
  df_list = []
  for i in tqdm(ks_list[:300]):
    ks_fundamental_df = stock.get_market_cap("20000101", "20231231", i, freq='m')
    ks_fundamental_df['symbol'] = i
    df_list.append(ks_fundamental_df)
  return df_list
# 각 데이터들의 per와 eps를 월 단위로 가져오기
ks_symbol_list = ks_symbol['amount'].to_list()
```

```python
ks_funda_list = bring_fundamental_info(ks_symbol_list)
```

```python
ks_funda_list
```

```python
df = pd.concat(ks_funda_list)
df.to_csv('./datasets/df3.csv')
```
