```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np
from pykrx import stock
kospi = yf.download('^KS11')
```

```python
from pykrx import stock
df = stock.get_index_ohlcv_by_date("1999-12-28","2024-01-05","1001","d")
df = df.sort_index()
```

```python
df_kospi = df[['종가']].rename(columns={'종가':'Close_KOSPI'})
df_kospi
```

## 2015 산타랠리
### 주가 상승 국면
    - 상승    127
    - 하락    121

```python
santa2015 = df_kospi.loc["2015-01-02":"2015-12-30"]
santa2015
```

```python
#다음날 종가 계산식
santa2015['Price_lag'] = santa2015['Close_KOSPI'].shift(-1)
```

```python
#주식 홀딩하는 경우 다음 날의 기대 수익
santa2015['expected_rtns'] = santa2015['Price_lag']-santa2015['Close_KOSPI']
santa2015
```

```python
#주가 변동율 계산식: 주식 holding 시 다음날 기대수익률
santa2015['Daily_rtn_rate'] = santa2015['expected_rtns']/santa2015['Close_KOSPI']
santa2015 
```

```python
santa2015['UPDOWN'] = [1 if santa2015['Daily_rtn_rate'].loc[date] > 0 else 0 for date in santa2015.index]
santa2015
```

```python
santa2015['Phase'] = ['상승' if x == 1 else '하락' for x in santa2015['UPDOWN']]
santa2015
```

```python
santa2015['Phase'].value_counts()
```
